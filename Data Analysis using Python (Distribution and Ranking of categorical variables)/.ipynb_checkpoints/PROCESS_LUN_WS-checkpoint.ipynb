```python
def unpivot_ws(excel_file, ws_name, counter_name):
    df = pd.ExcelFile(excel_file).parse(ws_name)
    timestamp_header = list(df.columns.values)[0]
    volume_header = 'LUN_Path'
    volume_column_headers = list(df.columns.values)[1:]
    volume_column = [x for x in volume_column_headers if counter_name in x]
    df = pd.melt(df, id_vars=[timestamp_header], value_vars=volume_column, var_name=volume_header, value_name=counter_name)
    df['LUN_Path'] = df['LUN_Path'].str.split(':').str[0]
    df['LUN_Path'] = df['LUN_Path'].str.replace('\'', "")
    return df
```

```python
read_ops = unpivot_ws(excel_file, '__lun', 'read_ops')
write_ops = unpivot_ws(excel_file, '__lun', 'write_ops')
read_data = unpivot_ws(excel_file, '__lun', 'read_data')
write_data = unpivot_ws(excel_file, '__lun', 'write_data')
avg_latency = unpivot_ws(excel_file, '__lun', 'avg_latency')
```
