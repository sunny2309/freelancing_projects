```python
def unpivot_ws(excel_file, ws_name, counter_name):
    df = pd.ExcelFile(excel_file).parse(ws_name)
    timestamp_header = list(df.columns.values)[0]
    volume_header = 'Volume_Name'
    volume_column_headers = list(df.columns.values)[1:]
    volume_column = [x for x in volume_column_headers if counter_name in x]
    df = pd.melt(df, id_vars=[timestamp_header], value_vars=volume_column, var_name=volume_header, value_name=counter_name)
    df['Volume_Name'] = df['Volume_Name'].str.split(':').str[0]
    df['Volume_Name'] = df['Volume_Name'].str.replace('\'', "")
    return df
```

```python
total_ops = unpivot_ws(excel_file, '__volume', 'total_ops')
read_data = unpivot_ws(excel_file, '__volume', 'read_data')
write_data = unpivot_ws(excel_file, '__volume', 'write_data')
avg_latency = unpivot_ws(excel_file, '__volume', 'avg_latency')
concurrency = unpivot_ws(excel_file, '__volume', 'concurrency')
parent_aggr = unpivot_ws(excel_file, '__volume', 'parent_aggr')
```
