**Detailed Business Problem**

Orkney Islands are one of the leaders in generating energy through renewable sources. Orkney Islands has around 35 wind turbines in working condition capable of generating 900 kW of power supply per turbine. Currently energy produced at Orkney by using these wind turbines easily handles demands of local households of Orkney as well as they are net energy exporter due to surplus energy after local usage. Orkney is connected to Great Britain by inter-connector of 40 MW capacity which limits transfer of energy between them. Orkney turbine owners are curtailing any energy which is produced after satisfying local household demands and transfer through inter-connector(40 MW) by setting set-point for controling power generation. This is the main concern raised by Orkney turbine owners as they feel that they can get more return on turbines investment if somehow energy which is getting curtailed due to limitation of inter-connector capacity can be used by Orkney households at discounted price for some purpose. As a Data Scientist, We have been asked to analyze household demand and turbine telemetry data to suggest whether there is any business case to invest in Demand Response (DR) Strategy which can help in utilizing these curtailed energy by local households. 

**Data:**

We have been provided with time-series data of power production by single turbine collected every 1 min(~1.06 Mn records) and average household demand data collected every 30 mins(~17.5 k records). Households data has information about average household demand (kW) and number of households. Turbine Telemetry has information about power generated by turbine (kW), set-point (kW) and wind speed (m/s). 

**Prime Objective:**

Main aim of doing this exercise is to find out whether there is case investing money in which helps all main parties involved in DR exercise in coming future:
* **Turbine Owners** - Will get return on curtailed energy usage.
* **DR Strategy Suppliers** - Will get return on implementation of DR strategy.
* **Individual Households** - Will get discounted energy which can be used for house heating which is currently done using fossil fuel.

**Approach:**

Approach followed to find solution to business problem has followed list of steps mentioned below. Whole analysis has been based on list of assumptions which were needed to make decisions about strategy suggestion and value calculations. Primary technologies used for doing in-depth analysis are python and its libraries like pandas, numpy, matplotlib, seaborn, datetime and collections. Whole analysis has been maintained as report in Jupyter Notebook with in-depth analysis with various calculations as well about DR strategies. This report can be used by other researchers as well to reproduce same results given same data. Analysis involves testing of various hypothesis about relation between various attributes of data which will helps us a lot in suggesting DR strategy and finding out solution to business problem. It takes on average 2-3 minutes to run notebook to reproduce results.Energy is considered curtailed if turbine has can produce more energy but has been limited by turbine owners by setting set-point somewhere below 900kw.

We have tried to answer below 3 questions while solving problem:
1. How much energy is currently curtailed annually across the Orkney Isles?
1. How much can this be reduced by different levels of DR penetration?
1. How many local households would need to be on our DR scheme in order to supply this level of DR?

**Steps:**


1. **Data Cleaning**

Data cleaning task consist of loading data, filtering based on year(2017), removing missing data rows, deriving new attributes from existing columns and resampling to match collection frequency. Turbine telemetry data has also been divided into curtailed, non-curtailed and turbine shutdown. New attributes like month, weekday name, weekday/weekends, season, hour of day based on datetime index of data. We have also resampled demand data from sampled at every 30 mins to every 1 mins to match it with frequency of turbine telemetry data.


2. **Exploratory Data Analysis**

EDA involves in-depth analysis with plotting about data based on various attributes generated as part of data cleaning process. It tries to find all possible patterns about data generation, demand and curtailment to find best DR strategy. EDA also involves testing of various hypothesis on household demand and turbine telemetry data to understand behaviour of data based on various derived attributes like season, months, time of day etc. EDA tries to find out pattern of data based on derived attributes for better decision making.

3. **DR Strategy Calculation**

DR calculation strategy involves finding out usage of curtailed enery by individual households. It involves in-depth calculations about curtailed enery available per house, DR penetration required to utilise curtailed energy, investment needed for implementing DR strategy (cost of smart meter etc.) and profit generated by DR strategy. Here, DR penetration refers to involvement of number of households in DR exercise(50% DR penetration refers to 50% households invovlement in DR). Various levels of DR penetration(10%,20%,50% &100%) has been tried and calculations are done for each strategy. Benchmark calculations are done for one turbine of island and then extended further to all turbines. All calculations are done making few assumptions about orkeny islands population, involvement in DR exercise etc.


**Conclusion:**
After doing in-depth analysis about data, trying variuos hypothesis and DR calculations, we suggested to invest in DR strategy as it was helping turbine owners generate more profit, helping individual households to use curtailed energy at discounted price and helping DR strategy owners to make money by implementing DR strategy on Orkeny Islands. Project was generally referred to as Heat Smart Orkney because currently households of Orkeny Islands uses fossil fuels for house heating purpose and will be using curtailed energy for same purpose which will reduce fuel(non-renrewable resource) poverty.
