```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import matplotlib.cm as cm

```

```python
# Example showing only Cyclical Strength Index plotted...

plt.figure(figsize = (20,10))

df['csi1'].plot(label = 'CSI Line 1')
df['csi2'].plot(style = '--', color = 'purple', label = 'CSI Line 2')

plt.legend()
```
