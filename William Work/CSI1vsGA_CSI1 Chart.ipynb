```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
import seaborn as sns
```

```python
df1 = pd.read_csv("df1.csv")

df1.head()
```

```python
def hexbin(df, grid_size=25):
    with plt.style.context("seaborn"):
        plt.figure(figsize=(10,8))

        plt.hexbin(x=df.csi1, y=df.ga_csi1,
                   C=df.next_day_close,
                   cmap="RdYlGn",
                   gridsize=grid_size,
                   #marginals=True
                  );

        #plt.xlim(min(df.csi1), max(df.csi1))
        #plt.ylim(min(df.ga_csi1), max(df.ga_csi1))

        plt.colorbar()

        plt.xlabel("CSI1")
        plt.ylabel("GA_CSI1")
        plt.title("CSI1 vs GA_CSI1 Contour Chart");
```

```python
hexbin(df1)
```

```python
hexbin(df1, 20)
```

```python
def contour1(df):
    X, Y = np.meshgrid(df.csi1, df.ga_csi1)
    Z = np.array([df.next_day_close.values.tolist()] * df.shape[0])

    norm = cm.colors.Normalize(vmax=abs(Z).max(), vmin=-abs(Z).max())

    plt.figure(figsize=(8,8))
    plt.contour(X,Y,Z,250,
               cmap="RdYlGn",
               );

    plt.colorbar()

    plt.xlabel("CSI1")
    plt.ylabel("GA_CSI1")
    plt.title("CSI1 vs GA_CSI1 Contour Chart");
```

```python
contour1(df1)
```

```python
def contour2(df):
    X, Y = np.meshgrid(df.csi1, df.ga_csi1)

    Z = np.repeat(df.next_day_close.values, df.shape[0]).reshape(df.shape[0], df.shape[0])

    norm = cm.colors.Normalize(vmax=abs(Z).max(), vmin=-abs(Z).max())

    plt.figure(figsize=(8,8))
    plt.contour(X,Y,Z,250,
               cmap="RdYlGn",
               );

    plt.colorbar()

    plt.xlabel("CSI1")
    plt.ylabel("GA_CSI1")
    plt.title("CSI1 vs GA_CSI1 Contour Chart");
```

```python
contour2(df1)
```
