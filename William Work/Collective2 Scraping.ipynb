```python
# Getting market cap for any ticker
from yahoo_fin import stock_info

def shares_out(ticker = 'MSFT'):
    
    shares_outstanding = stock_info.get_stats(ticker).set_index('Attribute').loc['Shares Outstanding 5']
    
    # Strip out M or B (for millions or billions of shares)
    if 'B' in shares_outstanding.values[0]:
        return float(shares_outstanding.values[0].strip('B')) * 1000000000
    
    elif 'M' in shares_outstanding.values[0]:
        return float(shares_outstanding.values[0].strip('M')) * 1000000
    
# Calculate market cap 
shares_out('TSLA') * stock_info.get_live_price('TSLA')
```
