```python
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import rc
from mpl_toolkits.axes_grid.axislines import Subplot

import math
from matplotlib.ticker import FuncFormatter
```

```python
df = pd.read_csv('stacked_bar_df.csv', index_col = 0)
```

```python
r = [x for x in range(0, len(list(df.index)), 1)]
```

```python
len(list(df.index))
```

```python
df = df.sort_values('occurences_total', ascending = False)

top_10 = [x for x in df['percent_top_10_occurences']]

the_rest = [x for x in df['percent_the_rest_occurences']]


mpl.rcParams['axes.spines.right'] = False
mpl.rcParams['axes.spines.top'] = False

colors = ['#c2c2f0','#ffb3e6']
colors2 = ['#f9bc86', '#a3acff']

plt.figure(figsize = (30,10))

barWidth = 0.85
names = tuple((list(df.index)))

ax = plt.bar(r, top_10, bottom = the_rest, label = 'Top 10 Titles', width = barWidth, color = colors2[0])
ax2 = plt.bar(r, the_rest, label = '"The Rest"', edgecolor='white', width = barWidth, color = colors2[1])

# Custom x axis
plt.xticks(r, names, fontsize = 25, rotation = 25)
plt.yticks(fontsize = 25)
plt.ylabel('Total Est. Viewership (TV Shows + Movies)', fontsize = 22)
plt.title('Concentration of 10 Most Popular Titles', fontsize=35)
plt.legend(fontsize = 25, title = '% Viewership Share', title_fontsize = '25') # prop={'size': 25},
plt.grid(False)
plt.subplots_adjust(left=0, bottom=0, right=.7, top=1, wspace=5, hspace=5)
plt.tight_layout()
plt.savefig('stacked_bar_top_occurences.png')
```

```python
ax = df.plot.bar(y=[ "occurences_the_rest", "occurences_top_10"], 
                 stacked=True,
                 figsize=(30,12), 
                 width=0.85,
                 rot=90, 
                 color=colors2[::-1], #["tomato", "lime"],
                )
handles,labels = ax.get_legend_handles_labels()
plt.legend(handles=handles[::-1],
           labels=["Top 10 Titles", '"The Rest"'], 
           fontsize = 25, 
           title = '% Viewership Share', title_fontsize = 25);

for i, (percent, tot) in enumerate(df[["percent_top_10", "occurences_total"]].values):
    plt.text(i, tot, "%.2f%%"%(percent*100), fontsize=22, ha="center", va="bottom")

plt.xticks(fontsize = 25)

def millions(x, pos):
    return '%1.1fM' % (x * 1e-6)

plt.yticks(fontsize = 25)
ax.yaxis.set_major_formatter(FuncFormatter(millions))
plt.ylabel('Total Est. Viewership\n(TV Shows + Movies)', fontsize = 25)

plt.title('Concentration of 10 Most Popular Titles', fontsize=35)
plt.subplots_adjust(left=0, bottom=0, right=.7, top=1, wspace=5, hspace=5)
plt.tight_layout()
plt.savefig('stacked_bar_top_occurences.png', dpi=200);
```
