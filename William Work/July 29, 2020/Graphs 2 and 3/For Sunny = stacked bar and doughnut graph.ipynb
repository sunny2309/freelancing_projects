```python
import pandas as pd
import datetime
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.patches import ConnectionPatch
from matplotlib.ticker import FuncFormatter
from matplotlib import gridspec
```

```python
df = pd.read_csv('occurences_prod_co.csv', index_col = 0)
df_results = pd.read_csv('viacom_top_10.csv', index_col = 2)
df['top_10_percent'] = df['top_10'] / df['occurences']
df['bottom_90_percent'] = df['bottom_90'] / df['occurences']
production_co = 'Viacom'
```

```python
df_results.sort_values('occurences', ascending = False, inplace = True)
```

```python
labels = list(df_results.index)
```

```python
sizes = [x for x in df_results['occurences']]

large = sizes[:int(len(sizes) / 2)]
small = sizes[int(len(sizes) / 2):]
reordered = large[::2] + small[::2] + large[1::2] + small[1::2]
```

```python
labels = []

for i in range(0, len(df_results), 1):

    labels.append(list(df_results[df_results['occurences'] == sizes[i]].index)[0])
```

```python
explode = []
mean = np.nanmean(df_results['occurences'])

for i in df_results['occurences']:
    
    if i > mean:
        explode.append(0.0)
    else:
        explode.append(0.0)
```

```python
def my_autopct(pct):
    return (str('%.2f' % pct)+'%') if pct > 2.5 else ''
```

```python
angle = 180 + float(sum(small[::2])) / sum(reordered) * 360

fig1, ax1 = plt.subplots(figsize = (10,10))
wedges, labels, autopct = ax1.pie(sizes, explode=explode, labels=labels, autopct=my_autopct,
        shadow=False, pctdistance=0.85, labeldistance=1.05, startangle = angle)

ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

#draw circle
centre_circle = plt.Circle((0,0),0.60,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

for i in range(0, len(labels), 1):
    text = labels[i].get_text()
    # Calculate percent 
#     percent = df_results.loc[text]['occurences'] / total
#     if percent > .025:
    labels[i].set_fontsize(15)
```

```python
spec = gridspec.GridSpec(ncols=6, nrows=1)

fig = plt.figure(figsize=(25,9))

bar_df = pd.DataFrame(df.loc[production_co]).T

ax1 = fig.add_subplot(spec[0])
ax1.spines["top"].set_visible(False)
ax1.spines["right"].set_visible(False)

ax1 = bar_df.plot.bar(y=["bottom_90","top_10",],
                       stacked=True,
                       width=0.9,
                       rot=25, 
                       color=['#a3acff', '#f9bc86'],
                       ax=ax1
                       )

plt.xticks(fontsize = 18)

def millions(x, pos):
    return '%1.1fM' % (x * 1e-6)

plt.yticks(fontsize = 18)
ax1.yaxis.set_major_formatter(FuncFormatter(millions))
plt.ylabel('Total Est. Viewership\n(TV Shows + Movies)', fontsize = 18)

handles,labels = ax1.get_legend_handles_labels()
#plt.legend([], [])
plt.legend(handles=handles[::-1],
           labels=["Top 10 Titles", '"The Rest"'], 
           loc="center right",
           fontsize = 12, 
           title = '% Viewership Share', title_fontsize = 12);

plt.title('Concentration of %s'%production_co, fontsize=18)

#### Percentage Annotations ###################
for i, (percent, tot) in enumerate(bar_df[["top_10_percent", "occurences"]].values):
    plt.text(i, tot, "%.2f%%"%(percent*100), fontsize=18, ha="center", va="bottom")
    

####################### Pie Chart #####################################    
ax2 = fig.add_subplot(spec[1:])


colors_gender = ['#c2c2f0','#ffb3e6', '#c2c2f0','#ffb3e6', '#c2c2f0','#ffb3e6', '#c2c2f0','#ffb3e6',
                 '#c2c2f0','#ffb3e6', '#c2c2f0','#ffb3e6', '#c2c2f0','#ffb3e6', '#c2c2f0','#ffb3e6',
                 '#c2c2f0','#ffb3e6', '#c2c2f0','#ffb3e6', '#c2c2f0','#ffb3e6', '#c2c2f0','#ffb3e6',
                 '#c2c2f0','#ffb3e6', '#c2c2f0','#ffb3e6', '#c2c2f0','#ffb3e6', '#c2c2f0','#ffb3e6',
                 '#c2c2f0','#ffb3e6']

angle = 180 + float(sum(small[::2])) / sum(reordered) * 360

def my_autopct(pct):
    return (str('%.2f' % pct)+'%') if pct > 2.5 else ''

wedges, labels, autopct = ax2.pie(df_results['occurences'].values, labels=df_results.index, autopct=my_autopct,
                                  wedgeprops=dict(width=0.5, edgecolor='w'),
                                  textprops=dict(size=15),
                                  shadow=False,
                                  pctdistance=0.75, 
                                  labeldistance=1.05, 
                                  startangle = angle,
                                  rotatelabels=True)

theta1, theta2 = ax2.patches[0].theta1, ax2.patches[0].theta2
center, r = ax2.patches[0].center, ax2.patches[0].r
x = r * np.cos(np.pi / 180 * theta2) + center[0]
y = np.sin(np.pi / 180 * theta2) + center[1]

######### Line 1 : Between Start of Top10 Bar and Lower Part of Pie Chart ###########################
con1 = ConnectionPatch(xyA=(0.45, df.loc[production_co]["bottom_90"]), xyB=(-0.5, -0.9), coordsA="data",
                      coordsB="data", axesA=ax1, axesB=ax2)
con1.set_color("black")
ax1.add_artist(con1)
con1.set_linewidth(3)

######### Line 1 : Between End of Top10 Bar and Upper Part of Pie Chart ###########################

con2 = ConnectionPatch(xyA=(0.45, df.loc[production_co]["occurences"]), xyB=(-0.6, 1.1), coordsA="data",
                      coordsB="data", axesA=ax1, axesB=ax2)
con2.set_color("black")
ax1.add_artist(con2)
con2.set_linewidth(3)

fig.subplots_adjust(wspace=0)

fig.tight_layout()
plt.savefig("graph2.png")
```
