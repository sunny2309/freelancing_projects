```python
import pandas as pd


import matplotlib
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from matplotlib.ticker import FuncFormatter
```

```python
df = pd.read_csv("scatter_plot.csv")
df
```

```python
grid = GridSpec(nrows=1, ncols=3)

fig = plt.figure(figsize=(15,5))

#colors = coolwarm(df["Avg Annual Ret"])
colors = ["royalblue", "skyblue", "slategrey", "darkorange"]

######################### Scatter PLot #########################
ax1 = fig.add_subplot(grid[:2])

ax1.scatter(x=df["Volatility"]*100, y=df["Avg Annual Ret"]*100, s=400, c=colors)

plt.xlabel("Annualized Volatility", fontdict={"size":15,})# "color":"gray"})
plt.ylabel("Annualized Return", fontdict={"size":15, })#"color":"gray"})
plt.title("Return to Volatility", fontdict={"size":20, "weight":"bold", "color":"dimgray", "fontname":"serif"}, )


def percent(x, pos):
        return '%1.1f%%' % (x)

ax1.xaxis.set_major_formatter(FuncFormatter(percent))
ax1.yaxis.set_major_formatter(FuncFormatter(percent))

ax1.tick_params(axis='both', which='both',length=0, pad=8)

plt.xticks(fontsize=15, ) #color="gray")
plt.yticks(fontsize=15, ) #color="gray")

x1, x2 = min(ax1.get_xticks()), max(ax1.get_xticks())
y1, y2 = min(ax1.get_yticks()), max(ax1.get_yticks())

plt.text(x1+0.25, y2-4.2, "Higher risk-adjusted\nreturn than S&P 500", fontsize=15)
plt.text(x2-1.4, y1+3.3, "Lower risk-adjusted\nreturn than S&P 500", fontsize=15)

plt.grid(True)

############### Bar Chart ##########################
ax2 = fig.add_subplot(grid[2])

ax2.barh(range(df.shape[0]), width=df["Sharpe Ratio"], height=0.4, color=colors)

plt.yticks(range(df.shape[0]), df["Portfolio"], fontsize=15)
ax2.yaxis.tick_right()

ax2.tick_params(axis='both', which='both', length=0, pad=8)

plt.title("Sharpe Ratio", fontdict={"size":20, "weight":"bold", "color":"dimgray",  "fontname":"serif"});


plt.tight_layout()

plt.savefig("scatter_plot1.png", dpi=130)
```
