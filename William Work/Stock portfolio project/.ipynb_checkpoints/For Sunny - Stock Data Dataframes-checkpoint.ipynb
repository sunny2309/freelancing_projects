```python
import pandas as pd
import datetime
import numpy as np
```

```python
# Import dataframes
df_pct_change = pd.read_csv('df_pct_change.csv', index_col = 0)
df_days_valid = pd.read_csv('df_count.csv', index_col = 0)
df_score = pd.read_csv('df_score.csv', index_col = 0)

# Convert Indexes to datetime format
df_pct_change.index = pd.to_datetime(df_pct_change.index)
df_days_valid.index = pd.to_datetime(df_count.index)
df_score.index = pd.to_datetime(df_score.index)
```

```python
# Example syntax: 
# Find all tickers that have been valid between 0 and 7 days... then from those valid tickers, find the ones with a score greater than 8.

date = '20200914'
date_dt = pd.to_datetime(date)
score_threshold = 7

# Step 1 - Filter for only tickers that have been valid between 0 and 7 days
columns = df_count.loc[date_dt].where(lambda x: (x > 0) & (x < 4)).dropna().sort_values().index

# Step 2 - From those valid tickers, find ones that have a score greater than 8
df_score[columns].loc[date].sort_values(ascending = False).where(lambda x: x > score_threshold).dropna()
```

```python
# Example syntax
# Create a portfolio where the valid tickers with the highest 7 scores will be included. Calculate cumulative sum for the portfolio

df_score_filtered = df_score.copy()

top = 7

for i, j in df_score.iterrows():
    
    # Get top 10 tickers for this day 
    top_tickers = df_score.loc[i].dropna().sort_values(ascending = False).index[:top].values.tolist()    
    
    for col in df_score.columns:
        
        if col not in top_tickers:
        
            df_score_filtered.loc[i, col] = None

# Calculate cumulative sum over the entire timeframe
df_top = df_score_filtered > -100 # Create a new dataframe with True/False where valid numbers exists
df_top = df_top.replace({True : np.nan})
df_top.update(df_pct_change, overwrite = False)
df_top = df_top.replace({False : None})
df_top.mean(axis = 1).cumsum() # Calculate cumulative sum of the portfolio
```

```python
max_tickers = 10
min_score = 4.0

valid_tickers = []

for dt in df_days_valid.index:
    
    ## Filter df_days_valid dataframe for current date
    days_valid_temp = pd.DataFrame(df_days_valid.loc[dt])
    days_valid_temp.columns = ["DaysValid"]
    
    ## Check for Valid Tickers Validness
    current_valid_tickers_df = days_valid_temp[days_valid_temp.index.isin(valid_tickers)]
    tickers_which_got_invalid_today = current_valid_tickers_df[current_valid_tickers_df.DaysValid < 1].index.tolist()
    
    ## This logic will remove ticker from valid_tickers which got invalid today.
    ## It'll remove them from valid_tickers so that space for new valid tickers to be added will be available.
    #print(tickers_which_got_invalid_today)
    for invalid_ticker in tickers_which_got_invalid_today:
        valid_tickers.remove(invalid_ticker)
    
    ## Taking only tickers which are not present in valid tickers list
    days_valid_temp = days_valid_temp[~days_valid_temp.index.isin(valid_tickers)]
    ## Filtering dataframe where days valid is more than 0 and sort it from lowest to highest
    days_valid_temp = days_valid_temp[days_valid_temp.DaysValid != 0].sort_values(by=["DaysValid"])
    ## Below Logic will keep first 20 rows with lowest days as valid.
    ## This is done so that in future when we combine it with scores and sort it according to score
    ## Then rows greater than 20 with more score does not get included as it might have more days.
    ## This can be changed if needed. Please let me know if this line is not needed.
    days_valid_temp = days_valid_temp[:20]
    
    ## If dataframe does not have any tickers with more than 0 days validity then continue to next date.
    ## Nothing can be added if dataframe is empty.
    if days_valid_temp.shape[0] == 0:
        print("Date: ", dt.date(), " Valid Tickers : ", valid_tickers, "\n")
        continue
        
    ## If dataframe has tickers which are valid for more than 0 days then filter score dataframe for this date.
    score_temp = pd.DataFrame(df_score.loc[dt])
    score_temp.columns = ["Score"]
    ## This will only keep rows where score is more than minimum score specified.
    score_temp = score_temp[score_temp.Score > min_score]
    
    ## Join days_valid_temp with score_temp on left so that only tickers in days_valid_temp stays in final dataframe.
    combined_df = days_valid_temp.join(score_temp, how="left")
    
    ## Sort dataframe according to score to take one with maximum score into consideration.
    ## Ascending=False will sort with more score first.
    combined_df = combined_df.sort_values(by=["Score"], ascending=False)
    combined_df = combined_df[combined_df.Score > min_score]
        
    ## Count space in valid tickers for tickers to be added. 
    ## If it already has max 10 tickers then below variable will be 0 hence nothing will be added.
    new_tickers_cnt = max_tickers - len(valid_tickers)
    
    ## Take new tickers from combined dataframe with most score 
    ## and according to space for no of tickers in valid tickers
    ## If new_tickers_cnt is 0 then this list will be empty and nothing will be added.
    ## Else tickers with new_tickers_cnt will be added to valid_tickers list.
    ## This logic will make sure that number of valid tickers always stays 10.
    new_tickers = combined_df.index.tolist()[:new_tickers_cnt]
    
    valid_tickers.extend(new_tickers)
    
    print("Date: ", dt.date(), " Valid Tickers : ", valid_tickers, "\n")
```
