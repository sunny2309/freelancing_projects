Explain in your own words how the Pearson correlation works and write down it's mathematical formulation. Can you think of an example where it fails (and visualization works)?
> * **ANS:** Pearson Correlation finds out coefficient which lets us know how one variable is related to another variable.

$$
Corr_{X,Y} = {\dfrac{cov(X,Y)} {\sigma_X\sigma_Y}}
$$

* {cov}  is the covariance
* ${\sigma _{X}}$ is the standard deviation of X
* ${\sigma _{Y}}$  is the standard deviation of Y

When there is no direct relation between two variables then pearson correlation fails and simple scatter plot might give better insights. Pearson correlation also does not work when there are lot of outliers as it can skew correlation coefficient.

> * What is the difference between a bar-chart and a histogram?
> * **ANS:** Bar chart represents quantity of each individual value of variable whereas histogram combines few variable quantities into one bin representing range of values.

> * I mention in the video that it's important to choose the right bin-size in histograms. But how do you do that? Do a Google search to find a criterion you like and explain it.
> * **ANS:** The Freedman-Diaconis rule is very robust and works well in practice. The bin-width is set to $h=2×IQR×n^{−1/3}$. So the number of bins is $(max−min)/h$, where n is the number of observations, max is the maximum value and min is the minimum value and IQR is inter quartile range.

> * Let's make a jitter-plot (that is, code up something like **Figure 2-1** from DAOST from scratch), but based on SF Police data. My hunch from inspecting the file is that the police-folks might be a little bit lazy in noting down the **exact** time down to the second. So choose a crime-type and a suitable time interval (somewhere between a month and 6 months depending on the crime-type) and create a jitter plot of the arrest times during a single hour (like 13-14, for example). So let time run on the $x$-axis and create vertical jitter.


```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy

%matplotlib inline

## Loading Data from https://data.sfgov.org/Public-Safety/Police-Department-Incident-Reports-2018-to-Present/wg3w-h783
sf_police_df = pd.read_csv('Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv')
sf_police_df.head()
```

```python
### Converting various datae time columns to proper format
sf_police_df['Date'] = pd.to_datetime(sf_police_df['Date'])
sf_police_df['Time'] = pd.to_timedelta([val+':00' for val in sf_police_df['Time']])
```

```python
sf_police_df['Category'].unique()
```

```python
## Taking 6 months data
six_months_data = sf_police_df[sf_police_df['Date'].isin(pd.date_range(start='6/1/2010', end='12/01/2010'))]
## Taking rows with crime type as drug offense
six_months_data = six_months_data[six_months_data['Category'] == 'DRUG/NARCOTIC']
## Taking crimes occurred between 7:00 AM to 8:00 AM
six_months_data = six_months_data[(six_months_data['Time'] > "07:00:00") & (six_months_data['Time'] < "08:00:00")]
## Grouping by Incident Time to get count of crime occurred at particular time.
six_months_data = six_months_data.groupby(by='Time').count()[['Date']].rename(columns={"Date":"Count"})
print(six_months_data.shape) ## Printing shape of resulting dataset.
six_months_data.head()
```

```python
plt.figure(figsize=(12,8))
## Plotting jitter plot
plt.scatter(range(six_months_data.shape[0]), [val + np.random.random() for val in six_months_data.Count], s=70, alpha=0.5)
plt.xticks(ticks=range(six_months_data.shape[0]), labels=[str(lab).split()[-1] for lab in six_months_data.index], rotation=90, );
plt.xlabel("Time");
plt.ylabel("Count of Crime")
plt.title('Jitter Plot of Drug Offences between 7:00 to 8:00');
```

> * Also I want to show you guys another interesting use of box plots. To get started, let's calculate another average for each focus-crime, namely what time of day the crime happens. So this time, the distribution we want to plot is the average time-of-day that a crime takes place. There are many ways to do this, but let me describe one way to do it. 
    * For datapoint, the only thing you care about is the time-of-day, so discard everything else.
    * You also have to deal with the fact that time is annoyingly not divided into nice units that go to 100 like many other numbers. I can think of two ways to deal with this.
      * For each time-of-day, simply encode it as seconds since midnight.
      * Or keep each whole hour, and convert the minute/second count to a percentage of an hour. So 10:15 $\rightarrow$ 10.25, 8:40 $\rightarrow$ 8.67, etc.
    * Now you can create box-plots to create an overview of *when various crimes occur*. Note that these plot have quite a different interpretation than ones we created in the previous exercise. Cool, right?  

```python
## Below function takes as input crime time and converts it to float as described above.
def convert_time(val):
    tm = str(val).split(' ')[-1]
    t1 = float(tm.split(':')[0])
    t2 = float(tm.split(':')[1])/ 60
    return t1+t2

sf_police_df_box_data["Time"] = [convert_time(val) for val in sf_police_df_box_data['Time']]
## Grouping by Crime Type and Incident Time.
sf_police_df_box_data2 = sf_police_df_box_data.groupby(by=['Category', "Time"]).count()[['IncidntNum']].rename(columns={'IncidntNum':'Count'})
sf_police_df_box_data2 = sf_police_df_box_data2.reset_index()
sf_police_df_box_data2.head()
```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(x="Category", y="Time",data=sf_police_df_box_data2)
plt.xticks(rotation=90)
plt.title("Box Plot for 15 Crime Types")
plt.xlabel("Incident Category")
plt.ylabel("Incident Time");
```

Yes. Output of this plot is quite different from previous one and help us decide when particular crime occurs most.
